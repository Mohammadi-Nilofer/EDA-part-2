# **Data Visualization**

* We can use box plot to detect the outliers in the data.



```python
df.plot(kind='box',subplots=True,layout=(3,5),figsize=(15,10))
```

**Observation:**

* By observing the above box plots, we can conclude that malic acid, magnesium and color intensity has more outliers.
* Alcalinity of ash and Proanthocyanins also has some outliers.

```python
df.Class.value_counts().plot(kind='bar',colormap='viridis')
```

Observation:

Class 2 is the highest and the class 3 is the lowest.



```python
sns.stripplot(y='Alcohol',x='Class',data=df)
```

Obervation:

* Class 2 has less amount of alcohol compared to class 1 and class 3



```python
sns.stripplot(y='Malic acid',x='Class',data=df)
```

Observation:

* Class 2 has higher level of malic acid.

```python
sns.swarmplot(y='Malic acid',x='Class',data=df)
```

Here, we can observe class2 and class 3 have higher level of malic acid and it has some outliers.

```python
sns.stripplot(y='Color intensity',x='Class',data=df)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

Observation:

* Highest positive correlation is between Flavanoids and Total phenols, followed by flavanoids and OD280/OD315 of diluted wines.
* Highest negative correlation is between Flavanoids and class, followed by Class and OD280/OD315 of diluted wines.
